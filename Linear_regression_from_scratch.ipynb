##Training and Testing the model

###( FIRST MODEL )

```python
X = df_NoNorm.drop(['Temperature (target)'], axis=1).values
y = df_NoNorm['Temperature (target)'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train = np.column_stack((np.ones(len(X_train)), X_train))
X_test = np.column_stack((np.ones(len(X_test)), X_test))
```

```python
learning_rate = 0.00001
n_iterations = 50 # because more than that will bring NaN values
beta = None

n_samples, n_features = X_train.shape
# Initialize beta with zeros
beta = np.zeros(n_features)

for i in range(n_iterations):
    # Calculate predictions
    print(beta)

    predictions = X_train.dot(beta)
    # Calculate residuals
    residuals = y_train - predictions
    # Calculate gradient
    gradient = -2/n_samples * X_train.T.dot(residuals)
    # Update beta using gradient
    beta = beta - learning_rate * gradient
    print("Iteration", i, " - Beta:", beta)
```

```python
y_pred = X_test.dot(beta)
```

####Calculate the Preformance Measures

R-squared

```python
from math import sqrt

#evaluate using r squared
rss = ((y_test - y_pred)**2).sum()
tss = ((y_test - y_test.mean())**2).sum()

r2_1 = 1 - (rss / tss)

print('R squared = ',r2_1)
```

MSE

```python
MSE_1 = 1/len(y_test)*(np.sum(np.subtract(y_test,y_pred)**2))
print("MSE=",MSE_1)
```

RSME

```python
rsme_1 = sqrt(MSE_1)
print("RSME=",rsme_1)
```

MAE

```python
MAE_1= np.mean(np.abs(y_test - y_pred))
print("MAE=",MAE_1)
```

```python
No_Norm_prof = pd.DataFrame({'real': y_test, 'predicted': y_pred})
No_Norm_prof.head(10)
```
